```python
import string
import re
from automata.fa.dfa import DFA
from visual_automata.fa.dfa import VisualDFA
```

```python
states={"Accept","Reject"}
alphabets = set(string.printable)
word=""
accept_states=[]
transition={
    "Accept":{},
    "Reject":{}
}
```

```python
print(alphabets)
```

```python
for c in alphabets:
    if c.islower():
      transition["Accept"][c]="Accept"
      transition["Reject"][c]="Reject"
    else:
      transition["Accept"][c]="Reject"
      transition["Reject"][c]="Reject"
```

```python
def dfa(word):
  state="Accept"
  for c in word:
    state=transition[state][c]
    if state=="Reject":
      print("Rejected")
      break
  if state=="Accept":
    print("Accepeted")
```

```python
text="this is sample text"
text2="Cat"
text3="cat"
dfa(text)
dfa(text2)
dfa(text3)
```

```python
initial_state="Accept"
final_state={"Reject"}
dfa = DFA(
    states=states,
    input_symbols=alphabets,
    transitions=transition,
    initial_state=initial_state,
    final_states=final_state,
)
```

```python
dfa = VisualDFA(
    states=states,
    input_symbols=alphabets,
    transitions=transition,
    initial_state=initial_state,
    final_states=final_state,
)
```

```python
dfa = VisualDFA(dfa)
```

```python
import json
import os

VOWELS = set("aeiou")

def load_irregulars(json_path="irregular-plurals.json"):
    if not os.path.exists(json_path):
        print(f"[Warning] {json_path} not found. Continuing with empty irregulars.")
        return {}
    with open(json_path, "r", encoding="utf-8") as fh:
        sing2pl = json.load(fh)
    pl2sing = {}
    for sing, pl in sing2pl.items():
        if isinstance(pl, str):
            pl2sing[pl.lower()] = sing.lower()
        elif isinstance(pl, list):
            for p in pl:
                if isinstance(p, str):
                    pl2sing[p.lower()] = sing.lower()
    return pl2sing

def analyze_word(w, pl2sing):
    w = w.strip()
    if not w:
        return None
    low = w.lower()

    if low in pl2sing:
        singular_root = pl2sing[low]
        return f"{w} = {singular_root}+N+PL"

    if low.endswith("ies"):
        stem_candidate = low[:-3] + "y"
        if len(low) >= 4:
            char_before_ies = low[-4]
            if char_before_ies not in VOWELS:
                return f"{w} = {stem_candidate}+N+PL"
            else:
                return f"{w} = Invalid Word"
        else:
            return f"{w} = Invalid Word"

    if low.endswith("es"):
        stem = low[:-2]
        if stem.endswith(("s", "x", "z")):
            return f"{w} = {stem}+N+PL"
        if stem.endswith(("ch", "sh")):
            return f"{w} = {stem}+N+PL"
        return f"{w} = Invalid Word"

    if low.endswith("s"):
        if len(low) >= 2:
            before = low[-2]
            last_two = low[-2:]
            if before in ("x", "z", "s"):
                return f"{w} = Invalid Word"
            if last_two in ("ch", "sh"):
                return f"{w} = Invalid Word"
            if low.endswith("ys"):
                if len(low) >= 3 and low[-3] not in VOWELS:
                    return f"{w} = Invalid Word"
                else:
                    root = low[:-1]
                    return f"{w} = {root}+N+PL" if root else f"{w} = Invalid Word"
            root = low[:-1]
            return f"{w} = {root}+N+PL" if root else f"{w} = Invalid Word"
        else:
            return f"{w} = Invalid Word"

    return f"{w} = {low}+N+SG"

def analyze_file(input_file, irregular_json, output_file):
    pl2sing = load_irregulars(irregular_json)
    results = []
    with open(input_file, "r", encoding="utf-8") as fh:
        for line in fh:
            token = line.strip()
            if not token:
                continue
            results.append(analyze_word(token, pl2sing))
    with open(output_file, "w", encoding="utf-8") as out_f:
        out_f.write("\n".join(results))
    print(f"Analysis complete. Results saved to {output_file}")

nouns_file = "brown_nouns.txt"
irregular_json = "irregular-plurals.json"
output_file = "analysis_results.txt"

analyze_file(nouns_file, irregular_json, output_file)
```
